## Preprocessing

### term

```python
loan_df['term'].unique()
```

```python
loan_df['term'] = loan_df['term'].str.replace(' 36 months', '36')
loan_df['term'] = loan_df['term'].str.replace(' 60 months', '60')

loan_df['term'] = pd.to_numeric(loan_df['term'])
loan_df['term'].unique()
```

### emp_length

```python
loan_df['emp_length'].unique()
```

```python
loan_df['emp_length'] = loan_df['emp_length'].str.replace('10+ years', '10', regex = False)
loan_df['emp_length'] = loan_df['emp_length'].str.replace('6 years', '6')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('4 years', '4')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('< 1 year', '0')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('2 years', '2')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('9 years', '9')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('5 years', '5')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('3 years', '3')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('7 years', '7')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('1 year', '1')
loan_df['emp_length'] = loan_df['emp_length'].str.replace('8 years', '8')
```

```python
loan_df['emp_length'] = pd.to_numeric(loan_df['emp_length'])
loan_df['emp_length'].unique()
```

### issue_d

```python
loan_df['issue_d'].unique()
```

```python
loan_df['issue_d'] = pd.to_datetime(loan_df['issue_d'], format = "%b-%Y")
loan_df['issue_d'].unique()
```

```python
# reference date 2020-01-01
mths_issue_d = list()
for date in loan_df['issue_d']:
    year = date.year
    month = date.month
    
    mths_issue_d.append(((2020-year)*12) + (1 - month))
loan_df['mths_issue_d'] = mths_issue_d
```

```python
loan_df['mths_issue_d'].unique()
```

### earliest_cr_line

```python
loan_df['earliest_cr_line'].isna().sum()
```

```python
loan_df['earliest_cr_line'].unique()
```

```python
loan_df['earliest_cr_line'] = pd.to_datetime(loan_df['earliest_cr_line'], format = "%b-%Y")
loan_df['earliest_cr_line'].unique()
```

```python
# reference date 2020-01-01
mths_earliest_cr_line = list()
for date in loan_df['earliest_cr_line']:
    year = date.year
    month = date.month
    
    mths_earliest_cr_line.append(((2020-year)*12) + (1 - month))
loan_df['mths_earliest_cr_line'] = mths_earliest_cr_line
```

```python
loan_df['mths_earliest_cr_line'].unique()
```

### Filling Missing Values

```python
loan_df.columns
```

```python
# print(loan_df['emp_length'].unique())
# 'verification_status_joint' in loan_df.columns
'initial_list_status' in loan_df.columns
```

```python
loan_df['annual_inc'].fillna(loan_df['annual_inc'].mean(), inplace=True)
loan_df['delinq_2yrs'].fillna(0, inplace=True)
loan_df['inq_last_6mths'].fillna(0, inplace=True)
loan_df['open_acc'].fillna(0, inplace=True)
loan_df['pub_rec'].fillna(0, inplace=True)
loan_df['total_acc'].fillna(0, inplace=True)
loan_df['total_rev_hi_lim'].fillna(0, inplace=True)
loan_df['emp_length'].fillna(0, inplace=True)
loan_df['mths_earliest_cr_line'].fillna(0, inplace=True)
```
